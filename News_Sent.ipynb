# Performing sentiment analysis

```python
import nltk
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer

sid = SentimentIntensityAnalyzer()
```

```python
df1['scores'] = df1['Headlines'].apply(lambda Processed_tweet: sid.polarity_scores(Processed_tweet))
df1.head()
```

```python
df1['compound']  = df1['scores'].apply(lambda score_dict: score_dict['compound'])
```

```python
df1.head(50)
```

```python
df2 = df1.iloc[:,[0,3]]
```

```python
df2
```

```python
News_Sent = df2.groupby("Date")["compound"].mean()
```
